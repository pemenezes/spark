```python
#importing necessary libraries
from pyspark.sql import SparkSession, Row, Column
from datetime import datetime, date
import pandas as pd
from pyspark.sql.functions import upper
```

```python
# starting a spark session | This is necessary to start a spark session always
spark = SparkSession.builder.getOrCreate()

# creating a df without the schema defined
df = spark.createDataFrame([
    Row(a=1, b=2, c= 'string1', d= date(2000, 1, 1), e= datetime(2000, 1, 1, 12, 0 )),
    Row(a=2, b=3, c='string2', d=date(2000, 2, 1), e=datetime(2000, 1, 2, 12, 0)),
    Row(a=4, b=5, c='string3', d=date(2000, 3, 1), e=datetime(2000, 1, 3, 12, 0))
])
```

```python
#showing the df
df
```

```python
#creating a df with the defined schema
df = spark.createDataFrame([
    Row(a=1, b=2, c= 'string1', d= date(2000, 1, 1), e= datetime(2000, 1, 1, 12, 0 )),
    Row(a=2, b=3, c='string2', d=date(2000, 2, 1), e=datetime(2000, 1, 2, 12, 0)),
    Row(a=4, b=5, c='string3', d=date(2000, 3, 1), e=datetime(2000, 1, 3, 12, 0))
], schema = 'a long, b long, c string, d date, e timestamp')
```

```python
df
```

```python
# transforming a pandas df into a spark df
pandas_df = pd.DataFrame({
    'a': [1, 2, 3],
    'b': [2., 3., 4.],
    'c': ['string1', 'string2', 'string3'],
    'd': [date(2000, 1, 1), date(2000, 2, 1), date(2000, 3, 1)],
    'e': [datetime(2000, 1, 1, 12, 0), datetime(2000, 1, 2, 12, 0), datetime(2000, 1, 3, 12, 0)]
})
df = spark.createDataFrame(pandas_df)
df
```

```python
#showing the df's
df.show()
df.printSchema()
```

```python
#showing only wanted number of rows in a df
df.show(1)
```

```python
#configuring the way we visalize the data when printing
#this makes more easy to visualize the data, in a way that looks like an excel
spark.conf.set('spark.sql.repl.eagerEval.enabled', True)
df
```

```python
#visualizing the name of the columns
df.columns
```

```python
# showing the describe
# it is important to know that the describe only shows it when used before .show()
df.describe().show()
```

```python
# we can select wich column do we want to see by selecting the columns
df.select('a','b').describe().show()
```

```python
df.collect()
```

```python
# we can transform as well a df that is in spark to pandas
df.toPandas()
df
```

```python
# we can easily create a new column in a df using the function withColumn()
df.withColumn('upper_c', upper(df.c))
```

```python
# we can also filter our df using the filter function
# this usage does not interfeer in the df itself, it only is used to show it once, but causes no changes in the df
df.filter(df.a == 1)
```

```python
# another df. It has its schema defined and we group the values
df = spark.createDataFrame([
    ['red', 'banana', 1, 10], ['blue', 'banana', 2, 20], ['red', 'carrot', 3, 30],
    ['blue', 'grape', 4, 40], ['red', 'carrot', 5, 50], ['black', 'carrot', 6, 60],
    ['red', 'banana', 7, 70], ['red', 'grape', 8, 80]], schema=['color', 'fruit', 'v1', 'v2'])
df.show()
```

```python
# we can use the groupby function too
# here we are grouping by color and showing the avg of the other columns in this groups
df.groupby('color').avg().show()
```

```python
# we can use a lot of diferent types of files to read and write using pyspark
# we will now use write and read a csv file
df.write.csv('test.csv', header = True)
spark.read.csv('test.csv', header = True)
```

```python
# we can as well create and read parquet files
df.write.parquet('test.parquet')
spark.read.parquet('test.parquet')
```

```python
# and we will lastely see how to crate and read ORC files
df.write.orc('test.orc')
spark.read.orc('test.orc')
```

```python
# working with SQL is easy using pyspark
df.createOrReplaceTempView('tableA')
spark.sql('select color from tableA')
```
